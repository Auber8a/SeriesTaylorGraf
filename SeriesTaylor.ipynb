Graficar polinomios de la serie de Taylor.

```python
import numpy as np
import matplotlib.pyplot as plt

def taylor_ln(x, n):
  result = 0
  for i in range(1, n + 1):
    result += ((-1)**(i+1)) * ((x-1)**i) / i
  return result


x_values = np.linspace(0.1, 2, 100)
n_terms = [1, 2, 3, 5, 10]

plt.figure(figsize=(10, 6))
plt.plot(x_values, np.log(x_values), label='ln(x)')

for n in n_terms:
    taylor_approximation = [taylor_ln(x, n) for x in x_values]
    plt.plot(x_values, taylor_approximation, label=f'Taylor n={n}')

plt.xlabel('x')
plt.ylabel('ln(x)')
plt.title('Aproximaci√≥n de ln(x) con la Serie de Taylor (x0=1)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def taylor_ln(x, n):
    result = 0
    for i in range(1, n + 1):
        result += ((-1)**(i+1)) * ((x-1)**i) / i
    return result

def taylor_cos(x, n):
    result = 0
    for i in range(n + 1):
        result += ((-1)**i) * (x**(2*i)) / np.math.factorial(2*i)
    return result

def taylor_1_over_1_minus_x(x, n):
    result = 0
    for i in range(n + 1):
        result += x**i
    return result


x_values_ln = np.linspace(0.1, 2, 100)
x_values_cos = np.linspace(-5, 5, 100)
x_values_1_over_1_minus_x = np.linspace(-0.9, 0.9, 100)


n_terms = [1, 2, 3, 5, 10]


plt.figure(figsize=(12, 8))

# ln(x)
plt.plot(x_values_ln, np.log(x_values_ln), label='ln(x)', color='blue')
for n in n_terms:
    taylor_approximation = [taylor_ln(x, n) for x in x_values_ln]
    plt.plot(x_values_ln, taylor_approximation, linestyle='--', label=f'Taylor ln(x) n={n}', alpha=0.7)

# cos(x)
plt.plot(x_values_cos, np.cos(x_values_cos), label='cos(x)', color='green')
for n in n_terms:
    taylor_approximation = [taylor_cos(x, n) for x in x_values_cos]
    plt.plot(x_values_cos, taylor_approximation, linestyle='--', label=f'Taylor cos(x) n={n}', alpha=0.7)

# 1/(1-x)
plt.plot(x_values_1_over_1_minus_x, 1 / (1 - x_values_1_over_1_minus_x), label='1/(1-x)', color='red')
for n in n_terms:
    taylor_approximation = [taylor_1_over_1_minus_x(x, n) for x in x_values_1_over_1_minus_x]
    plt.plot(x_values_1_over_1_minus_x, taylor_approximation, linestyle='--', label=f'Taylor 1/(1-x) n={n}', alpha=0.7)


plt.xlabel('x')
plt.ylabel('Valor de la funci√≥n')
plt.title('Aproximaci√≥n de ln(x), cos(x) y 1/(1-x) con la Serie de Taylor')
plt.legend(loc='upper right', fontsize='small', ncol=2)
plt.grid(True)
plt.ylim(-5, 5)


plt.show()
```
